```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
print(x)
```

```python
def lossFn(x):
    return (x-5)**2
```

```python
loss = lossFn(x)
```

```python
plt.style.use("seaborn")
plt.plot(loss)
plt.show()
```

```python
def gradDesc():
    x_ = 0
    learning_rate = 0.1
    print(x_)
    plt.plot(loss)
    for i in range(30):
        grad = 2*(x_-5)

        x_ = x_ - learning_rate*grad
        l = lossFn(x_)
        print("X_ : %.3f Loss: %.3f "%(x_, l))
        plt.scatter(x_, l)
```

```python
gradDesc()
```
