# Data Visulisation Intro

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
plt.style.use("seaborn")
```

```python
X = np.arange(10)
y1 = X**2
print(y1)
y2 = 2*X + 3
print(y2)
```

```python
plt.plot(X, y1, color = 'red', label = 'apple', marker = 'o')
plt.plot(X, y2, color = 'green', label = 'kiwi', linestyle = 'dashed', marker = '*')
plt.xlabel = 'time'
plt.ylabel = 'price'
plt.title = 'price vs time'
plt.show()
```

```python
prices = np.array([1, 2, 3, 4])**2
print(prices)
plt.plot(prices)
plt.show()
```

### Scatter Plot

```python
plt.figure(figsize=(2, 2))
plt.scatter(X, y1, color = 'red', label = 'apple', marker = 'o')
plt.scatter(X, y2, color = 'green', label = 'kiwi', linestyle = 'dashed', marker = '*')
plt.xlabel = 'time'
plt.ylabel = 'price'
plt.title = 'scatter - price vs time'
plt.legend()
plt.show()
```

### Bar Graph

```python
plt.style.use('dark_background')
x_coordinates = np.array([0, 1, 2])*2
plt.bar(x_coordinates, [10, 20, 15], width=0.5, label= 'current year',tick_label= ["Gold", "Platinum", "Diamond"], color = 'blue')
plt.bar(x_coordinates+0.5, [20, 10, 12], width=0.5, label='next year', color = 'orange')
plt.title = "Metal Price Comparison"
plt.xlabel = 'Metal'
plt.ylabel = 'Price'
plt.ylim(0, 40)
plt.legend()
plt.show()
```

### Pie Chart

```python
plt.style.use("seaborn")
subjects = "Maths", "Chem", "Phy", "Eng"
weightage = 20, 10, 15, 5
plt.pie(weightage, labels=subjects, explode=(0, 0, 0.1, 0), autopct='%1.1f%%', shadow= True)
```

### Histogram

```python
xsn = np.random.randn(100)
sigma = 5
u = 60
X = np.round(xsn * sigma + u)
X2 = np.round(xsn*7 + 40)
print(X)
```

```python
plt.hist(X, alpha= 0.8, label= 'maths')
plt.hist(X2, alpha=0.8, label='physics')
plt.xlable = "Marks Range"
plt.ylabel = "Prob/Freq of Students"
plt.legend()
plt.show()
```

# DV Challenge

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\VS\Coding-Blocks-ML\Data Visualisation\movie_metadata.csv")
```

```python
df.head(10)
```

```python
df.columns
```

```python
data = list(df['movie_title'])
#print(data)
print(data[0])
```

```python
freq_data = {}

for title in data:
    length = len(title)

    if freq_data.get(length) is None:
        freq_data[length] =  1

    else:
        freq_data[length] += 1
```

```python
print(freq_data)
```

```python
X = np.array(list(freq_data.keys()))
Y = np.array(list(freq_data.values()))
```

```python
plt.scatter(X, Y)
plt.xlabel = 'length of movie title'
plt.ylabel = 'freq count'
plt.title = 'movie title visualization'
plt.show()
```
