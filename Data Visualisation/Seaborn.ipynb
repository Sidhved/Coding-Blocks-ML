# Seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data= tips, estimator=np.std)
```

```python
sns.countplot(x= 'sex', data=tips)
```

```python
sns.countplot(x='time', data=tips)
```

```python
sns.boxplot(x= 'day', y= 'total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
sns.displot(x='total_bill', data= tips)
```

```python
sns.displot(tips['total_bill'], kde= False, bins = 30)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x= 'total_bill', y= 'tip', data=tips, kind='hex')
```

```python
sns.pairplot(data=tips, hue='sex')
```

```python
tipc = tips.corr()
tipc
```

```python
sns.heatmap(tipc, annot= True)
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
flight_pivot = flights.pivot_table(index= 'month', columns= 'year', values= 'passengers')
flight_pivot
```

```python
sns.heatmap(flight_pivot, cmap='coolwarm')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic_pivot = titanic.pivot_table(index='sex', columns='class', values='survived')
```

```python
sns.heatmap(titanic_pivot)
```
