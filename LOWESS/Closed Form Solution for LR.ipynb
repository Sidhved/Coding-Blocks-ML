```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
X, Y = make_regression(n_samples= 400, n_features=1, n_informative=1, noise=1.8, random_state=11)

Y = Y.reshape((-1, 1))
print(X.shape)
print(Y.shape)
```

```python
X = (X-X.mean())/X.std()
plt.figure()
plt.scatter(X, Y)
plt.title("Normalised Data")
plt.show()
```

```python
ones = np.ones((X.shape[0], 1))
X_ = np.hstack((X, ones))
print(X_.shape)
print(X_[:5, :])
```

```python
def predict(X, theta):
    return np.dot(X, theta)

def getTheta(X, Y):
    Y = np.mat(Y)
    f = np.dot(X.T, X)
    s = np.dot(X.T, Y)

    theta = np.linalg.pinv(f)*s
    return theta
```

```python
theta = getTheta(X_, Y)
print(theta)
```

```python
X = (X-X.mean())/X.std()
plt.figure()
plt.scatter(X, Y)
plt.plot(X, predict(X_, theta), color='red')
plt.title("Normalised Data")
plt.legend()
plt.show()
```
