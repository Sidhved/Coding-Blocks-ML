```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5) ** 2
print(X, Y)
```

### Visualisation

```python
plt.style.use('seaborn')
plt.plot(X, Y)
plt.xlabel = ('X')
plt.ylabel = ('F(X)')
plt.show()
```

```python
x = 0
lr = 0.1
error = []
#50 steps downhill
for i in range(50):
    grad = 2*(x-5)
    x= x-lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x, y)
    print(x)
plt.show()
```

```python
#Plot the values of error
plt.plot(error)
```
